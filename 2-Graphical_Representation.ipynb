```python
#Importing Dataset
import pandas as pd
dataset = pd.read_csv(r"data_cleaned.csv", encoding= 'iso-8859-1')
```

```python
#Checking if data is properly read
dataset.head()
```

```python
# Numerical Summary
temp = dataset.describe()
temp.to_csv('stat.csv', encoding='iso-8859-1')
```

# Boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
datasetNew = dataset[['tradeTime','followers','totalPrice','price', 'square','communityAverage']]

#Plotting boxplot
ax = plt.boxplot(datasetNew)
sns.show()
```

### Boxplots for each column to determine outliers

```python
import matplotlib.pyplot as plt

#Plotting boxplots for each column
for column in dataset:
    plt.boxplot(dataset[column])
    plt.title(column)
    plt.show()
```

# Histograms

```python
#Histograms for each column
dataset.hist(bins=50, figsize=(20,15))
plt.show()
```

# Scatter Plots
### Scatter Plot 1

```python
x=dataset['Lat']
y=dataset['Lng']

#Scatter Plots between coordinates for better understanding of data
scttr = plt.scatter(x, y, alpha=0.4, \
    s=dataset['totalPrice']/100, label='Price', \
    c=dataset['district'], cmap=plt.get_cmap('jet'))
plt.colorbar(scttr)
plt.legend()
plt.show()
```

### Scatter Plot 2

```python
x=dataset['Lat']
y=dataset['Lng']
scttr = plt.scatter(x, y, alpha=0.4, \
    c=dataset['totalPrice'], cmap=plt.get_cmap('jet'))
plt.colorbar(scttr)
plt.show()
```

```python
x=dataset['communityAverage']
y=dataset['totalPrice']

# Histogram between community Average and Total Price for better understanding of data

plt.hist(x, weights=y/10**2)
plt.xlabel('Community Average')
plt.ylabel('Total price')
plt.title('Community Average to the Total price')
plt.show()
```

# Correlations

```python
#Checking correlations between columns
corr = plt.subplots(figsize=(25, 25))
sns.heatmap(dataset.corr(), annot = True, linewidth = .5, fmt = ".3f")
plt.show()
```

# Probplots

```python
import pylab 
import scipy.stats as stats

# Creating probplots for each column
for column in dataset:
    stats.probplot(dataset[column], dist="norm", plot=pylab)
    plt.title(column)
    plt.show()
```
