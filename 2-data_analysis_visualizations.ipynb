# Data Analysis

```python
#Importing Dataset
import pandas as pd
dataset = pd.read_csv(r"data_cleaned.csv", encoding= 'iso-8859-1')
```

```python
#Checking if data is properly read
dataset.head()
```

# Descriptive Analysis

### Numerical Stats

```python
dataset.describe()
```

## Find Outliers
### Boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
datasetNew = dataset[['tradeTime','followers','totalPrice','price', 'square','communityAverage']]

#Plotting boxplot
ax = sns.boxplot (data=datasetNew)
plt.show()
```

### Boxplots for all variables to determine outliers

```python
import matplotlib.pyplot as plt

dataset2 = dataset.drop('floorType', axis = 1)
#Plotting boxplots for all variables
for column in dataset2:
    plt.boxplot(dataset2[column])
    plt.title(column)
    plt.show()
```

### Histograms

```python
#Histograms for all variables
dataset.hist(bins=50, figsize=(20,15))
plt.show()
```

## Scatter Plots
### Scatter Plot 1

```python
x=dataset['Lat']
y=dataset['Lng']

#Scatter Plots between coordinates for better understanding of data
scttr = plt.scatter(x, y, alpha=0.4, \
    s=dataset['totalPrice']/100, label='Price', \
    c=dataset['district'], cmap=plt.get_cmap('jet'))
plt.colorbar(scttr)
plt.show()
```

### Scatter Plot 2

```python
x=dataset['Lat']
y=dataset['Lng']
scttr = plt.scatter(x, y, alpha=0.4, \
    c=dataset['totalPrice'], cmap=plt.get_cmap('jet'))
plt.colorbar(scttr)
plt.show()
```

### Histogram for totalPrice and communityAverage variable

```python
x=dataset['communityAverage']
y=dataset['totalPrice']

# Histogram between community Average and Total Price for better understanding of data

plt.hist(x, weights=y/10**2)
plt.xlabel('Community Average')
plt.ylabel('Total price')
plt.title('Community Average to the Total price')
plt.show()
```

# Exploratory Analysis
## Checking Correlations

### Numerical Representation

```python
dataset.corr()
```

### Graphical Representation

```python
#Checking correlations between columns
corr = plt.subplots(figsize=(25, 25))
sns.heatmap(dataset.corr(), annot = True, linewidth = .5, fmt = ".3f")
plt.show()
```

### Probplots for all variables

```python
import pylab 
import scipy.stats as stats

# Creating probplots for each column
for column in dataset2:
    stats.probplot(dataset2[column], dist="norm", plot=pylab)
    plt.title(column)
    plt.show()
```
